```python
### Jaikumar Gurobi Implementation with binary contract
%pip install gurobipy
import gurobipy as gp
from gurobipy import GRB
import pandas as pd


m = gp.Model("Jaikumar with contract")
extra_amount=5000

#defining variables
var_index = ['15','84']
Xvar = m.addVars(var_index, name="X")
Yvar = m.addVar(vtype=GRB.BINARY, name="Y")

#defining the objective
m.setObjective((179-88)*Xvar['15'] + (124-74)*Xvar['84'] +(182-88)*Yvar*extra_amount, GRB.MAXIMIZE)


#defining the constraints
spinning_time = m.addConstr((1/30)*(Xvar['15']+Yvar*extra_amount) + (1/60)*Xvar['84'] <= 300,
name='spinning')
drawtwisting_time = m.addConstr((1/15)*(Xvar['15']+Yvar*extra_amount) + (1/12)*Xvar['84'] <= 1600,
name='drawtwisting')
min_prod_qty_15 = m.addConstr(Xvar['15']+Yvar*extra_amount >= 3000, name='min_kg_15')
min_prod_qty_84 = m.addConstr(Xvar['84'] >= 6000, name='min_kg_84')
max_prod_qty_15 = m.addConstr(Xvar['15']+Yvar*extra_amount <= 10000, name='max_kg_15')
max_prod_qty_84 = m.addConstr(Xvar['84'] <= 15000, name='max_kg_84')

#solving the model
m.write('jaikumar_contract.lp')
m.optimize()
print(m.x)
```
